SOAL No. 2

```python
q1 = whr['Social support'].quantile(0.25)
q3 = whr['Social support'].quantile(0.75)

iqr = q3-q1

lower = q1-1.5*iqr
upper = q3+1.5*iqr

print('Lower Bound :', lower)
print('Upper Bound :', upper)
```

```python
outliers = whr[(whr['Social support']>upper) | (whr['Social support']<lower)]
outliers.head()
```

```python
outliers.to_excel('Nomor Dua_Outliers.xlsx', index=False)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=whr, x=whr['Social support'])
plt.title('Social Support Boxplot')
plt.xlabel('Social Support')
plt.grid(True)
plt.show()
```

SOAL No. 5

```python
cscore = whr[whr['Score'] > whr['Score'].mean()]
cscore.head()
```

```python
cscore.to_excel('Nomor Lima_Score.xlsx', index=False)
```

```python
pd.DataFrame(cscore[['Country or region','GDP per capita']]).head()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,6))
sns.barplot(x= cscore['GDP per capita'].head(5), y=cscore['Country or region'].head(5))
plt.xlabel('GDP per capita')
plt.ylabel('Country or region')
plt.title('Country with The Highest GDP')
plt.grid(True)
plt.show()
```

```python
pd.DataFrame(cscore[['Country or region','Perceptions of corruption']]).tail()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,6))
sns.barplot(x= cscore['Perceptions of corruption'].tail(5), y=cscore['Country or region'].tail(5))
plt.xlabel('Perceptions of corruption')
plt.ylabel('Country or region')
plt.title('Country with The Lowest Perceptions of Corruption')
plt.grid(True)
plt.show()
```

```python
pd.DataFrame(cscore[['Country or region','Generosity']]).tail()
```
